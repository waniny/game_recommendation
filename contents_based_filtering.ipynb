```python
from ast import literal_eval
import pandas as pd
import numpy as np
import warnings
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import re 
df = pd.read_csv('게임데이터_종합.csv')
df.head(3)
```

```python
df = df.dropna(axis=0)
df
```

```python
df["발매년"] = df["발매일"].str[-4:]
df = df.astype({'발매년':int})
df = df[df["발매년"]>=2000]
df
```

```python
df = df.drop(['발매년'],axis='columns')
df
```

```python
df['장르'] = df['장르'].apply(literal_eval)
df['태그'] = df['태그'].apply(literal_eval)
df[['장르','태그']].head(2)
```

```python
df['장르'] = df['장르'].apply(lambda x : (' ').join(x))
df['태그'] = df['태그'].apply(lambda x : (' ').join(x))
df
```

```python
C = df['선호도(%)'].mean()

# 백분위수 
m =df['리뷰수'].quantile(0.6)

def weighted_score_average(record):
    v = record['리뷰수']
    R = record['선호도(%)']
    
    return ( (v/(v+m)) * R) + ( (m/(m+v)) * C)
    
df['가중평점'] = df.apply(weighted_score_average, axis=1)
```

```python
count_vector = CountVectorizer(ngram_range=(1,3))

c_vector_genres = count_vector.fit_transform(df['장르'])
c_vector_tags = count_vector.fit_transform(df['태그'])
```

```python
#코사인 유사도를(장르) 구한 벡터 저장
genre_c_sim = cosine_similarity(c_vector_genres, c_vector_genres).argsort()[:,::-1]
# tag_c_sim = cosine_similarity(c_vector_tags,c_vector_tags).argsort()[:,::-1]
genre_c_sim
```

```python
#가중평점을 고려한 game 추천  

def get_recommend_game_list(df, game_name , top=10):
    
    #특정 게임과 비슷한 게임추천을 위한 '특정 게임'정보 뽑기
    target_game_index  = df[df['게임명'] == game_name].index.values
    
    #코사인 장르 유사도 중 비슷한 코사인 장르 유사도 정보 뽑아내기
    sim_genre_index = genre_c_sim[target_game_index, :51].reshape(-1)
    #본인 제외
    sim_genre_index = sim_genre_index[sim_genre_index != target_game_index]
    
    #장르 유사도가 비슷한 것들 중 코사인 태그 유사도를 구한 벡터 저장
    tag_c_sim = cosine_similarity(c_vector_tags,c_vector_tags)[target_game_index, sim_genre_index]

    #코사인유사도(태그)비슷한 것 정보 뽑아내기 
    sim_index = tag_c_sim.argsort().reshape(-1)[:top]
    
    
    #data frame으로 만들고 retrun
    result = df.iloc[sim_index].sort_values('가중평점', ascending=False )[:10]
    return result
```

```python
selected_game = df[df['게임명'] == "Little Nightmares"]
selected_game = selected_game.drop(['발매일','선호도(%)','리뷰수'], axis = 'columns')
selected_game
```

```python
df2 = get_recommend_game_list(df, "Little Nightmares" , top=10)
recommended_game_list = df2.drop(['발매일','선호도(%)','리뷰수'], axis = 'columns')
recommended_game_list
```
